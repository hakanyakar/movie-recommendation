```python
import numpy as np
import pandas as pd

ratings_data = pd.read_csv("ratings.csv")#getting ratings from csv
ratings_data.head()
```

```python
movie_names = pd.read_csv("movies.csv")#getting movies from csv
movie_names.head()
```

```python
movie_names['title'] = movie_names['title'].str.split('(', 1).str[0].str.strip()#Splitting years from movie title
```

```python
movie_data = pd.merge(ratings_data, movie_names, on='movieId')#merge two data frames
movie_data.head()
```

```python
movie_data.groupby('title')['rating'].mean().head()#calculate the mean of the rating for each movie.
```

```python
movie_data.groupby('title')['rating'].mean().sort_values(ascending=False).head()#sort the ratings in the descending order of their average ratings
```

```python
movie_data.groupby('title')['rating'].count().sort_values(ascending=False).head()#total number of ratings for a movie
```

```python
ratings_mean_count = pd.DataFrame(movie_data.groupby('title')['rating'].mean())#add the average rating of each movie to this dataframe
ratings_mean_count['rating_counts'] = pd.DataFrame(movie_data.groupby('title')['rating'].count())#add the number of ratings for a movie to the ratings_mean_count dataframe
ratings_mean_count.head()
```

```python
user_movie_rating = movie_data.pivot_table(index='userId', columns='title', values='rating')#create the matrix of movie titles and corresponding user ratings
user_movie_rating.head()
```

```python
film_name = input()
```

```python
film_name_ratings = user_movie_rating[film_name]#user ratings for selected film by user
film_name_ratings.head()
```

```python
movies_like_film_name = user_movie_rating.corrwith(film_name_ratings)#Correlation ratings and movie

corr_film_name = pd.DataFrame(movies_like_film_name, columns=['Correlation'])
corr_film_name.dropna(inplace=True)#remove NaN 
corr_film_name.head()
```

```python
corr_film_name.sort_values('Correlation', ascending=False).head(10)#sort the movies in descending order of correlation
```

```python
corr_film_name = corr_film_name.join(ratings_mean_count['rating_counts'])#join ratings_count column
corr_film_name.head()
```

```python
corr_film_name[corr_film_name ['rating_counts']>50].sort_values('Correlation', ascending=False).head()#sort movies descending correlation and more than 50 ratings
```
